```python
dummy_variable_1 = pd.get_dummies(df["genero"])
dummy_variable_1.head()
```

```python
dummy_variable_1.rename(columns={'AD':'Action,Adventure', 'ADC':'Action,Adventure,Drama', 'ADF':'Action,Adventure,Fantasy', 'ADM':'Action,Adventure,Mystery', 'ADS':'Action,Adventure,Sci-Fi', 'ABD':'Action,Biography,Drama', 'ACC':'Action,Comedy,Crime'}, inplace=True)
dummy_variable_1.head()
```

```python
df = pd.concat([df, dummy_variable_1], axis=1)
df.drop("genero", axis = 1, inplace=True)
df.head
```
